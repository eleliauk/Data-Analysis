# Matplotlib

What is it?
- a low level graph plotting library used for visualizing data
- open source and freely available
- mostly written in Python

Let's go through some examples to see the basics of using Matplotlib

First, need to import matplotlib.pyplot. Usual one gives a short new name ("plt") to avoid a lot of typing later

```python
import matplotlib.pyplot as plt
import numpy as np

print("done")
```

- The top "window" where things are drawn is created using plt.figure(). But one can also let Python do that automatically 

- One can can have multiple independent figures

- each figure can contain multiple Axes objects

- Axes define the area where our plots will be drawn. Each Axes object has an x-axis and a y-axis

```python

fig = plt.figure()
ax = fig.add_subplot(111) 
```

- if nothing happens, need to call plt.show()
- so far, only an empty figure with default size and default axis range

```python
plt.show()
```

- Axes object belongs to a particular figure 
- contains axis and has attributes that can be used to annotate plot
- A subplot is an axes on a grid system contolled by arguments in e.g. add_subplot(). Need to check documentation on definition of the grid system
- can also create Axes ourselves to put it anywhere on the figure without grid system
- we also set axis labels for x and y axis, the range of the x and y axis and a title for the plot

```python
fig = plt.figure()

# Creating subplot/axes
ax = fig.add_subplot(111)

# Setting plot title
ax.set_title('My plot title')

# Setting X-axis and Y-axis limits
ax.set_xlim([0, 10])
ax.set_ylim([-5, 5])

# Setting X-axis and Y-axis labels
ax.set_ylabel('My y-axis label')
ax.set_xlabel('My x-axis label')

# Showing the plot
plt.show()
```

- Matplotlib objects usually have explicit setters to control attributes. 
- Will often need to consult Matplotlib reference guide to see available arguments
- we can also set many attributes when creating the axis to avoid some typing
- same result as above

```python
fig = plt.figure()

# Creating subplot/axes
ax = fig.add_subplot(111)

# Setting title and axes properties
ax.set(title='An Axes Title', xlim=[0, 10], ylim=[-5, 5], ylabel='My y-axis label', xlabel='My x-axis label')

plt.show()
```

- as often the case in Python, there are several different ways of achieving the same thing
- most methods of the Axes object are also available  method in the pyplot module.
- they will act on the current plotting axes object
- e.g. one can call plt.xlable('X-Axis Label') to set the label of X-axis (remember we set plt as alias for pyplot)
- this will just call ax.set_xlabel('X-Axis Label') for the currently active axes

```python
# Creating subplots, setting title and axes labels using `pyplot`
plt.subplots()
plt.title('Plot using pyplot')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.show()
```

- figure object can contain multiple axes
- these can be created usign plt.subplots()
- going through subplots() will put axes on a grid system (as said above)

Example of 2x2 subplots:

```python
# Creating subplots with 2 rows and 2 columns
fig, axes = plt.subplots(nrows=2, ncols=2)
plt.show()
```

- NOTE: The axes object returned by subplots(nrows=2, ncols=2) is a 2D-numpy array of axes objects
- each element corresponds to one of the 4 subplots in the figuree
- can access the 4 plots using array indices 

Example how to use the array of axes:

```python
# Create a figure with four subplots and shared axes
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)
axes[0, 0].set(title='Upper Left',ylabel='Y-Axis Label')
axes[0, 1].set(title='Upper Right')
axes[1, 0].set(title='Lower Left')
axes[1, 1].set(title='Lower Right', xlabel = 'X-Axis Label')
plt.show()
```

Ok, time to actually plot some data

- many different plotting functions available
- will only show simple example
- will find more using Matplotlib reference as neededd
- plot() method from pyplot is good starting point
- syntax is plot([x], y, [fmt], data=None, **kwargs):

- easiest case is plotting points or lines using set of x and y coordinates
- The optional parameter fmt can be used for basic formatting, e.g.,  color, marker type, and style 

- To plot a line, we provide coordinates along X and Y axes separately

```python
x = [1.3, 2.9, 3.1, 4.7, 5.6, 6.5, 7.4, 8.8, 9.2, 10]
y = [95, 42, 69, 11, 49, 32, 74, 62, 25, 64]

# Plot lists 'x' and 'y'
plt.plot(x, y)

# Plot axes labels and show the plot
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

- can even just provide y coordinates
- index will be used as substitute of x coordinates

Example:

```python
# Defining 'y' coordinates
y = [95, 42, 69, 11, 49, 32, 74, 62, 25, 32]

# Plot list 'y'
plt.plot(y)

# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in List Y')
plt.show()
```

- the blue line above is the default Python style and color. It is ugly. If I see a plot like this I know that zero effort went into making the plot look good
- The optional parameter fmt in the plot method can be used for defining basic formatting like color, marker, and line-style. It is a shortcut string notation consisting of color, marker, and line:

fmt = '[color][marker][line]'

```python
# Plot line with green color
plt.plot(y, 'g')

# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in List Y')
plt.show()
```

- markers can be added using the same notation as shown below

```python
# Plot continuous green line with circle markers
plt.plot(y, 'go-')

# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in List Y')
plt.show()
```

- we may want to use different colors for markers and line
- need to plot the same data twice, with different format options

```python
# Plot list 'y'
plt.plot(y, 'g')

# Plot red circle markers
plt.plot(y, 'ro')

# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in List Y')
plt.show()
```

- using the same approach we can also plot two sets of data on the same plot

Example:

```python
# Define two lists
y = [95, 42, 69, 11, 49, 32, 74, 62, 25, 32]
y2 = [35, 52, 96, 77, 36, 66, 50, 12, 35, 63]
print(type(y))

# Plot lists and show them
plt.plot(y, 'go-')
plt.plot(y2, 'b*--')

# Plot axes labels and show the plot
plt.xlabel('Index Values')
plt.ylabel('Elements in Lists')
plt.show()
```

- plot can be used to plot many different sequential Python data structures, e.g. list, NumPy arrays, pandas series, etc. 

Example of plotting a NumPy one dimensional array:

```python
# Importing NumPy library
import numpy as np

# Drawing 30 samples from a standard normal distribution into an array 'arr'
arr = np.random.normal(size=30)
print(arr)

# Plotting 'arr' with dashed line-style and * markers
plt.plot(arr, color='teal', marker='*', linestyle='dashed')
```

- so far looked at 1D plots
- we will often want to look at correlations between variables using 2-D plots
- simplest version is a so-called scatter plot: plt.scatter()
- In simplest use, x and y parameters are array-like sequential data structures

```python
# Data points for scatter plot
x = [1.3, 2.9, 3.1, 4.7, 5.6, 6.5, 7.4, 8.8, 9.2, 10]
y = [95, 42, 69, 11, 49, 32, 74, 62, 25, 32]

color = np.random.rand(10)
size = np.random.randint(50, 100, 10)

# Creating a scatter plot
plt.scatter(x, y, s=size, c=color)

# Labeling the plot and showing it
plt.xlabel('Values from list x')
plt.ylabel('Values from list y')
plt.show()
```

- can plot points with information encoded in color and size as shown above
- can be omitted (and often is for scientific data looking for correlations)

Example:

```python
# Creating a scatter plot without color and same
plt.scatter(x,y)

# Labeling the plot and showing it
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

# Histograms

- histograms are essential tools in many kinds of statistical data analysis
- contain the frequency distribution of values in data 
- usual frequency of values is counted in finite ranges ("bins") of x-axis variable
- i.e., divide the entire range into a series of intervals 
- count how many values fall into each interval
- Bins are consecutive and non-overlapping
- bins are on x-axis,  y-axis shows the frequency of each bin
- created using the hist() method of pyplot

Example:

```python
# Data values for creating a histogram
y = [95, 42, 69, 11, 49, 32, 74, 62, 25, 32]

# Creating a histogram
plt.hist(y)
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

- hist() method did everything for us: creating bins, segregating each value to the corresponding bin, counting and  plotting
- can specify number of bins, or use default
- can easily understand output

- more interesting example
- use np.random.normal to provide 10000 random numbers, distributed as a normal distribution, with mean 0 and standard deviation 1
- use hist() to create histogram

```python
# Creating an array
array = np.random.normal(0, 1, 10000)

# Creating a histogram
plt.hist(array)
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

- as expected this looks like a normal distrbution (bell curve)
- this blue, solid color histogram is the default in Matplotlib. If I see a plot like this, I know NO effort was made to make the plot visually attractive
- can provide arugments to change color and type of plotting, such as bar, barstacked, step and stepfilled. 

Example: 

```python
# Creating an array
array = np.random.normal(0, 1, 10000)

# Creating a histogram and plotting it
plt.hist(array, color='purple', histtype='step')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

- This looks a little bit better, but a lot of work is needed to make this "publication quality"
- Can change orientation of plot from vertical to horizontal (99% of plots we publish are horizontal)

```python
# Creating an array
array = np.random.normal(0, 1, 10000)

# Creating a histogram and plotting it
plt.hist(array, color='teal', orientation='horizontal')
plt.xlabel('Frequency')
plt.ylabel('Bins')
plt.show()
```

- Coming back to 2D plots. The simplest way to plot 2D correlation information are scatter plots. 
- Scatter points become useless when there is a lot of data and the dots fill the whole plot
- Many interesting options available to plot 2D data
- below is an example using hist2D and making a so-called contour plot

```python
from matplotlib.colors import LogNorm
import numpy as np

# create some data for plotting
num_particles = 1000
eta = np.zeros(1000)
phi = np.zeros(1000)
for i in range(1000):
    eta[i] = np.random.rand()*2.0 - 1.0
    phi[i] = np.random.rand()*3.14159*2

deta = []
dphi = []
for i in range(1000):
    for j in range(i+1,1000):
        d1 = eta[i] - eta[j]
        d2 = phi[i] - phi[j]
        deta.append(d1)
        dphi.append(d2)
        
# separate the histogramming and plotting steps
# hist2d will do the binning, picking the axis range, counting and plotting for us, returning the histogram data as counts and bins
plt.subplot(121)
counts,xbins,ybins,image = plt.hist2d(deta,dphi,bins=100
                                      ,norm=LogNorm()
                                      , cmap = plt.cm.rainbow)

#now take the counts and bins and use contour() function to plot a nice looking contour plot

plt.colorbar()
plt.subplot(122)
plt.contour(counts,extent=[xbins[0],xbins[-1],ybins[0],ybins[-1]],
    linewidths=3, cmap = plt.cm.rainbow, levels = [1,5,10,25,50,70,80,100])
```
